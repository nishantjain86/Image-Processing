```python
import matplotlib.pyplot as plt
import cv2
print(cv2.__version__)
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1 = cv2.imread('/content/drive/MyDrive/JUIT_Courses/DIP using Python/images/a1.jpg')
if img1 is None:
  print("Canot read the image")
else:
  print("Sucessful")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('/content/drive/MyDrive/JUIT_Courses/DIP using Python/images/a2.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img3 = cv2.imread('/content/drive/MyDrive/JUIT_Courses/DIP using Python/images/a3.jpg')
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
plt.imshow(img3)
```

```python
#Making the shape of both the images same
img1= cv2.resize(img1,(1000,1000))
img2= cv2.resize(img2,(1000,1000))
img3= cv2.resize(img3,(1000,1000))
plt.imshow(img2)
```

```python
#using cv2 addition: If the addition result is greater than 255 (incase of uitn8) , then the pixel value will be automatically be converted into 255
img4 = cv2.add(img1,img2)
plt.imshow(img4)
```

```python
#using numpy addition: If the addition result is greater than 255 (incase of uitn8) , then the pixel value = addition result % 256 that is 300%256 = 44
img5 = img1+img2
plt.imshow(img5)
```

```python
# 1st argumnet: input first image
# 2nd argumnet: weight of first image
# 3rd argumnet: input second image
# 4th argumnet: weight of second image
# 5th argument: epsilon
dst = cv2.addWeighted(img1,0.5,img2,0.5,0)
plt.imshow(dst)
```

```python
img6 = cv2.subtract(img2,img3)
plt.imshow(img6)
```
