**1. Drawing lines on image using OpenCV Function.**

```python
# First argument: input image
# Second argument: top-left corner
# Third argument: bottom-right corners
# Fourth argument: Color of the shape
# Fifth argument: Thickness of the line. If -1 is passed, then color will be filled in the closed figures

cv.line(img1,(0,0),(511,511),(255,0,0),15)
plt.imshow(img1)
```

**2. Drawing Rectangles on image using OpenCV Function.**

```python
# First argument: input image
# Second argument: top-left corner
# Third argument: bottom-right corners
# Fourth argument: Color of the shape
# Fifth argument: Thickness of the line. If -1 is passed, then color will be filled in the closed figures
cv.rectangle(img1,(384,0),(510,128),(0,255,0),30)
plt.imshow(img1)
```

**3. Drawing Ellipses on image using OpenCV Function.**

```python
# First argument: input image
# Second argument: center location of ellipse
# Third argument: major axis length, and minor axis length of ellipse
# Fourth argument: angle of rotation of ellipse in anti-clockwise direction.
# Fifth and sixth argument: startAngle and endAngle denotes the starting and ending of ellipse arc measured in clockwise direction from major axis
# Seventh argument: COlor
# Eigth Argument: Thickness of the ellipse. If -1 is passed, then color will be filled in the closed figures
cv.ellipse(img1,(256,256),(300,150),30,0,180,(255,0,255),-1)
plt.imshow(img1)
```

**4. Drawing Circles on image using OpenCV Function.**

```python
# First argument: input image
# Second argument: center location of ellipse
# Third argument: Radius
# Fourth argument: COlor
# Fifth Argument: Thickness of the ellipse. If -1 is passed, then color will be filled in the closed figures

cv.circle(img1,(447,63), 63, (0,0,255), -1)
plt.imshow(img1)
```

```python
# First argument: input image
# Second argument: center location of ellipse
# Third argument: Radius
# Fourth argument: COlor
# Fifth Argument: Thickness of the ellipse. If -1 is passed, then color will be filled in the closed figures
img2 = img1.copy()
cv.circle(img2,(750,750), 63, (0,0,255), -1)
plt.imshow(img1)
plt.figure()
plt.imshow(img2)
```

**5. Writing a text on image using OpenCV  function**

```python
# 1st  argument: image
# 2nd argument: Text data that you want to write
# 3rd argument: Bottom left coordinate from where you want put text
# 4th argument: Font type
# 5th argument: Font Scale (specifies the size of font)
# 6th argument: Color
# 7th argument: Thickness
# 8th argument: lineType = cv.LINE_AA 

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img1,'OpenCV',(10,500), font, 4,(255,255,255),2,cv.LINE_AA)
plt.imshow(img1)
```
